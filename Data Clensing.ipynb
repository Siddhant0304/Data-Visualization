```python
#Importing the packages required for visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the data from the CSV file and copying it into a data frame
Hospital_beds_df = pd.read_csv("Changed.csv")

#Print and check the Dataframe
Hospital_beds_df
```

```python
#checking column names
print(Hospital_beds_df.columns)
```

```python
#Displaying information of the every column in dataset
Hospital_beds_df.info()
```

```python
#Displaying sum of empty rows in each column
Hospital_beds_df.isna().sum()
```

```python
#Cleansing the data
print(Hospital_beds_df['Patients Currently Hospitalized'].mean())
print(Hospital_beds_df['Patients Newly Admitted'].mean())
print(Hospital_beds_df['Patients Positive After Admission'].mean())
print(Hospital_beds_df['Patients Discharged'].mean())
print(Hospital_beds_df['Patients Currently in ICU'].mean())
print(Hospital_beds_df['Patients Currently ICU Intubated'].mean())
print(Hospital_beds_df['Patients Expired'].mean())
print(Hospital_beds_df['Total Staffed Beds'].mode()[0])
print(Hospital_beds_df['Total Staffed Beds Currently Available'].mean())
print(Hospital_beds_df['Total Staffed ICU Beds'].mean())
print(Hospital_beds_df['Total Staffed ICU Beds Currently Available'].mean())
print(Hospital_beds_df['Patients Age Less Than 1 Year'].mean())
print(Hospital_beds_df['Patients Age 1 To 4 Years'].mean())
print(Hospital_beds_df['Patients Age 5 to 19 Years'].mean())
print(Hospital_beds_df['Patients Age 20 to 44 Years'].mean())
print(Hospital_beds_df['Patients Age 45 to 54 Years'].mean())
print(Hospital_beds_df['Patients Age 55 to 64 Years'].mean())
print(Hospital_beds_df['Patients Age 65 to 74 Years'].mean())
print(Hospital_beds_df['Patients Age 75 to 84 Years'].mean())
print(Hospital_beds_df['Patients Age Greater Than 85 Years'].mean())
```

```python
#For filling numerical data
num = ['Patients Currently Hospitalized','Patients Newly Admitted','Patients Positive After Admission','Patients Discharged','Patients Currently in ICU','Patients Currently ICU Intubated','Patients Expired','Total Staffed Beds Currently Available','Total Staffed ICU Beds','Total Staffed ICU Beds Currently Available','Patients Age Less Than 1 Year','Patients Age 1 To 4 Years','Patients Age 5 to 19 Years','Patients Age 20 to 44 Years','Patients Age 45 to 54 Years','Patients Age 55 to 64 Years','Patients Age 65 to 74 Years','Patients Age 75 to 84 Years','Patients Age Greater Than 85 Years']

for column in num:
    Hospital_beds_df[column].fillna(round(Hospital_beds_df[column].mean(),0), inplace=True)
```

```python
#For filling categorical data
cat = ['Total Staffed Beds']
for column in cat:
    Hospital_beds_df[column].fillna(Hospital_beds_df[column].mode()[0], inplace=True)
```

```python
#Check the cleansed data
Hospital_beds_df.isna().sum()
```
